# Solving differential equations

Solving equation $\displaystyle\frac{dy}{dx} = 2x + 1$ (with $F(0) = 0$) using Euler method.

```python
import numpy as np
import matplotlib.pyplot as plt

def findMean(numbers):
    sum = 0
    j = 0
    for i in numbers:
        sum += i
        j += 1
    
    return sum / j

h = 0.25
x = {}
derivative = {}
x['h=0.25'] = np.arange(0, 5.5, h)
derivative['h=0.25'] = 2 * x['h=0.25'] + 1

x['h=0.5'] = np.arange(0, 5.5, h * 2)
derivative['h=0.5'] = 2 * x['h=0.5'] + 1

y_exact = {}
y_exact['h=0.25'] = x['h=0.25'] ** 2 + x['h=0.25']
y_exact['h=0.5'] = x['h=0.5'] ** 2 + x['h=0.5']

y_euler = {}
y_euler['h=0.25'] = []
y_euler['h=0.25'].append(0)
for i in range (len(x['h=0.25']) - 1):
    y_euler['h=0.25'].append( y_euler['h=0.25'][i] + h * derivative['h=0.25'][i] )
    
y_euler['h=0.5'] = []
y_euler['h=0.5'].append(0)
for i in range (len(x['h=0.5']) - 1):
    y_euler['h=0.5'].append( y_euler['h=0.5'][i] + h * 2 * derivative['h=0.5'][i] )

square_diff = {}
square_diff['h=0.25'] = (y_euler['h=0.25'] - y_exact['h=0.25']) ** 2
square_diff['h=0.5'] = (y_euler['h=0.5'] - y_exact['h=0.5']) ** 2
    
plt.plot(x['h=0.25'], y_euler['h=0.25'])
plt.plot(x['h=0.25'], y_exact['h=0.25'])
plt.plot(x['h=0.5'], y_euler['h=0.5'])
# plt.scatter(x['h=0.25'], y_euler['h=0.25'], s=5)
# plt.scatter(x['h=0.25'], y_exact['h=0.25'], s=5)
# plt.scatter(x['h=0.5'], y_euler['h=0.5'], s=5)
plt.legend(['Euler h=0.25', 'Exact', 'Euler h=0.5'])
plt.title('Using Euler method')
plt.show()

print('Mean of squares of uncertainties when h = 0.25:', findMean(square_diff['h=0.25']))
print('Mean of squares of uncertainties when h = 0.5:', findMean(square_diff['h=0.5']))
```

Solving the same equation using midpoint method.

```python
def findMean(numbers):
    sum = 0
    j = 0
    for i in numbers:
        sum += i
        j += 1
    
    return sum / j

h = 0.25
x = {}
derivative = {}
x['h=0.25'] = np.arange(0, 5.5, h)
derivative['h=0.25'] = 2 * ( x['h=0.25'] + h/2 ) + 1

x['h=0.5'] = np.arange(0, 5.5, h * 2)
derivative['h=0.5'] = 2 * ( x['h=0.5'] + h ) + 1

y_exact = {}
y_exact['h=0.25'] = x['h=0.25'] ** 2 + x['h=0.25']
y_exact['h=0.5'] = x['h=0.5'] ** 2 + x['h=0.5']

y_midpoint = {}
y_midpoint['h=0.25'] = []
y_midpoint['h=0.25'].append(0)
for i in range (len(x['h=0.25']) - 1):
    y_midpoint['h=0.25'].append( y_midpoint['h=0.25'][i] + h * derivative['h=0.25'][i] )
    
y_midpoint['h=0.5'] = []
y_midpoint['h=0.5'].append(0)
for i in range (len(x['h=0.5']) - 1):
    y_midpoint['h=0.5'].append( y_midpoint['h=0.5'][i] + h * 2 * derivative['h=0.5'][i] )

square_diff = {}
square_diff['h=0.25'] = (y_midpoint['h=0.25'] - y_exact['h=0.25']) ** 2
square_diff['h=0.5'] = (y_midpoint['h=0.5'] - y_exact['h=0.5']) ** 2
    
# plt.plot(x['h=0.25'], y_midpoint['h=0.25'])
# plt.plot(x['h=0.25'], y_exact['h=0.25'])
# plt.plot(x['h=0.5'], y_midpoint['h=0.5'])
plt.scatter(x['h=0.25'], y_midpoint['h=0.25'], s=5)
plt.scatter(x['h=0.25'], y_exact['h=0.25'], s=5)
plt.scatter(x['h=0.5'], y_midpoint['h=0.5'], s=5)
plt.legend(['Midpoint h=0.25', 'Exact', 'Midpoint h=0.5'])
plt.title('Using midpoint method')
plt.show()

print('Mean of squares of uncertainties when h = 0.25:', findMean(square_diff['h=0.25']))
print('Mean of squares of uncertainties when h = 0.5:', findMean(square_diff['h=0.5']))
```

## Normal distribution

This is the plot of the [standard normal distribution](https://en.wikipedia.org/wiki/Normal_distribution#Standard_normal_distribution) ($\mu = 0, \, \sigma = 1$).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4, 4, 0.001)
y = (1 / np.sqrt(2*np.pi)) * np.exp(-(x**2)/2)

plt.plot(x, y)
plt.title('Standard normal distribution')
plt.ylabel('Probability density')
plt.show()
```

Plot of the [Cumulative Distribution function](https://en.wikipedia.org/wiki/Normal_distribution#Cumulative_distribution_function).

```python
h = 0.001
x = np.arange(-4, 4, h)
derivative = (1 / np.sqrt(2*np.pi)) * np.exp(-((x+h/2)**2)/2)

integral = []
integral.append(0)
for i in range (len(x) - 1):
    integral.append( integral[i] + h * derivative[i] )
    
plt.plot(x, integral)
plt.title('Cumulative distribution function')
plt.ylabel('Probability of Z being <x')
plt.show()

x1 = 1.5
x1_index = round((x1 + 4) * 1000)
x2 = 2.5
x2_index = round((x2 + 4) * 1000)

print('(' + str(x1) + ', ' + str(integral[x1_index]) + ')')
print('(' + str(x2) + ', ' + str(integral[x2_index]) + ')')
print('The probability that a number (Z) comes in range from ' + str(x1) + ' to ' + str(x2) + ' is ' + str(integral[x2_index] - integral[x1_index]))
```
